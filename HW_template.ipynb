```
# This section will import the appropriate packages

# Math packages
import numpy as np  # https://numpy.org/

# Plotting packages
import matplotlib.pyplot as plt  # https://matplotlib.org/index.html
```

```
# Here we will define the sinusoid function with the following inputs
# time line                 - time_line
# amplotide of the signal   - amplitude
# frequency of the signal   - frequency
# phase of the signal       - phase

def sinusoid(time_line, amplitude, frequency, phase):
    signal  = amplitude * np.sin( 2 * np.pi * frequency * time_line + phase)

    return signal
```

```
# This section will be the main part of your code that will do all the necessary manipulation and required calculations

# Comment the lines where you do complex computations and also if you use functions from other packages that were disucussed in class/office hours

t_start = 0.0  # start of the time line 
t_end   = 10.0 # end of the time line
dt      = 0.01  # time step

time_line = np.arange(t_start, t_end, dt) # creates an array of numbers from t_start to t_end with time step dt 

# characteristics of the first sinusoid
amplitude_1 = 2.0
frequency_1 = 0.5
phase_1     = np.pi / 4 

# characteristics of the second sinusoid
amplitude_2 = 1.0
frequency_2 = 1.0
phase_2     = np.pi 

# creating the signals using the functions sinusoid
sinusoid_1 = sinusoid(time_line, amplitude_1, frequency_1, phase_1) # first sinusoid
sinusoid_2 = sinusoid(time_line, amplitude_2, frequency_2, phase_2) # first sinusoid
```

```
# In this section we will plot the signals on a typical xy graph
# when plotting something, make sure that you have a TITLE, the AXES are WELL LABELED (with UNITS if available) 
# and the information you're plotting is CLEAR and LEGIBLE
# always include a LEGEND if you're plotting multiple things

plt.figure ( figsize = ( 20 , 12 ) )

plt.plot(time_line, sinusoid_1, color = 'red',    linewidth = 4, label = 'sin$_1$')  # plot command x-axis, y-axis
plt.plot(time_line, sinusoid_2, color = 'blue',   linewidth = 4, label = 'sin$_2$')  # plot command x-axis, y-axis

plt.legend(fontsize = 20) # plots the legend using the labels from above

# Plot title
plt.title('Sinusoidal Signals\nAmplitude$_1$ = '+str(amplitude_1)+'  Frequency$_1$ = '+str(frequency_1)+' Hz  Phase$_1$ = '+str( round(phase_1,2) )+'\nAmplitude$_2$ = '+str(amplitude_2)+'  Frequency$_2$ = '+str(frequency_2)+' Hz  Phase$_2$ = '+str(round(phase_2,2))+'\n', fontsize = 25)
# str(x) converts x to a string
# round(x,n) rounds x to n sig figs 

# Plot Axes Lable 
plt.ylabel('Amplitude', fontsize=25) 
plt.xlabel('time (s)' , fontsize=25)

# Feel free to addapt this template to your style of plotting, however keep in mind that we have to go through your code and UNDERSTAND WHAT YOU'RE DOING   


plt.show()
```
